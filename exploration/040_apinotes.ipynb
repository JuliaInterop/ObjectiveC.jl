Explore `apinotes` files in the SDK. 
=====================================
- These will help clean up naming of types and function in generated pacakges
- References
  - https://github.com/apple/swift/blob/main/apinotes/README.md
  - https://github.com/apple/swift/search?q=apinotes


## Find the SDK directory

```julia
sdk_dir = `xcrun --show-sdk-path` |> open |> readchomp 
```

```julia
using .Iterators
Base.filter(f) = x -> filter(f, x)
(r::Regex)(s) = occursin(r, s)
head(n::Int=10)  = x->take(x,n)
head(x) = head()(x)
```

```julia
apinotes = `find $sdk_dir -name "*.apinotes"` |> open |> readlines |> collect
```

```julia
apimap = Dict(((chopsuffix(splitpath(f)[end], ".apinotes"), f) for f in apinotes))
## TODO: Load in dataframe and explore file naming convention
```
